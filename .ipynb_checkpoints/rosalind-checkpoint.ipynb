### COUNTING DNA NUCLEOTIDES

```python
dna_string = input((""))
```

```python
A = "A"
C = "C"
G = "G"
T = "T"

counta = dna_string.count(A)
countc = dna_string.count(C)
countg = dna_string.count(G)
countt = dna_string.count(T)

print(counta,countc,countg,countt)
```

### Complementing a strand of DNA

```python
import pandas as pd

dna_string = input((""))
```

```python
def dna_complementar(seq):
    complementar = ""
    for nucleotideo in seq:
        if nucleotideo == "A":
            complementar += "T"
        elif nucleotideo == "T":
            complementar += "A"
        elif nucleotideo == "G":
            complementar += "C"
        else:
            complementar += "G"
    return complementar

complementar = dna_complementar(dna_string)

complementar_reversa = complementar[::-1]

complementar_reversa
```
